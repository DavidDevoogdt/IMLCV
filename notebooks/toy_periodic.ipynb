```python
from pathlib import Path

from IMLCV.base.rounds import Rounds
from IMLCV.examples.example_systems import toy_periodic_phase_trans
from IMLCV.scheme import Scheme

folder = Path("perovskites")/"toy_periodic_2"

if folder.exists():
    scheme = Scheme( rounds= Rounds.create(folder=folder, copy=False, new_folder=False))

else:
    md,refs = toy_periodic_phase_trans()
    

    scheme = Scheme.from_refs(mde=md, refs=sum(*refs), folder=folder,steps=1000,)
```

```python
steps=2000

chunk_size=100


n=15

max_bias=150 * kjmol
samples_per_bin=200
min_samples_per_bin=5

n_max=2*n
```

```python
scheme.inner_loop(
    steps=steps,
    n=n,
    n_max_fes=n_max,
    convergence_kl=1e-4,
    chunk_size=100,
    eps_umbrella=0.10,
    max_bias=max_bias,
    samples_per_bin=samples_per_bin,
    min_samples_per_bin=min_samples_per_bin,
)
```
