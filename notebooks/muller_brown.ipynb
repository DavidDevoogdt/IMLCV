```python
from pathlib import Path

from IMLCV.base.rounds import Rounds
from IMLCV.examples.example_systems import toy_1d
from IMLCV.scheme import Scheme

folder = Path("perovskites") / "toy_muller_brown_2"

if folder.exists():
    scheme = Scheme(rounds=Rounds.create(folder=folder, copy=False, new_folder=False))

else:
    md, refs, fes_0 = toy_1d()

    scheme = Scheme.from_refs(
        mde=md,
        refs=sum(*refs),
        folder=folder,
        steps=1000,
    )
```

```python
from IMLCV.base.UnitsConstants import boltzmann, kelvin, kjmol

300 * kelvin * boltzmann / kjmol
```

```python
steps = 1000

chunk_size = 100

max_bias = 150 * kjmol
samples_per_bin = 20
min_samples_per_bin = 5
```

```python
scheme.inner_loop(
    steps=steps,
    n=8,
    convergence_kl=1e-4,
    chunk_size=100,
    eps_umbrella=0.10,
    max_bias=max_bias,
    samples_per_bin=samples_per_bin,
    min_samples_per_bin=min_samples_per_bin,
)
```
